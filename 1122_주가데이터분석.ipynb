```python
# prompt: 초기 자금은 10억원이다 골든 크로스 이후 주식을 매일 100주씩 매입하다가 데드크로스 이후에서 매일 100주씩 파는 방식은 2020.1.1부터 현재까지 수행하면 이익인가 손실인가?

# 이전 코드에서 필요한 부분만 가져와서 수정했습니다.
import pandas as pd
from pykrx import stock

# 삼성전자 티커 심볼
ticker = '005930'  # company_info['삼성전자']를 직접 사용하는 것보다 명시적으로 ticker를 지정하는 것이 좋습니다.

# 2020년 1월 1일부터 현재까지의 주식 정보 가져오기
df = stock.get_market_ohlcv_by_date("20200101", "20240401", ticker)

# 10일, 25일 이동평균 계산
df['10일'] = df['종가'].rolling(window=10).mean()
df['25일'] = df['종가'].rolling(window=25).mean()

# 초기 자금
initial_money = 1000000000

# 주식 보유량
stock_count = 0

# 거래 내역 (데이터프레임 형태로 저장)
transactions = pd.DataFrame(columns=['Date', 'Action', 'Price', 'Quantity', 'Total'])


# 골든크로스, 데드크로스 판단 및 거래
golden_cross = False
for i in range(25, len(df)):
    if df['10일'][i-1] < df['25일'][i-1] and df['10일'][i] > df['25일'][i] : # 골든 크로스 발생
        golden_cross = True

    if golden_cross and initial_money >= df['종가'][i] * 100:
        stock_count += 100
        initial_money -= df['종가'][i] * 100
        transactions = pd.concat([transactions, pd.DataFrame({'Date': df.index[i], 'Action': 'Buy', 'Price': df['종가'][i], 'Quantity': 100, 'Total': df['종가'][i] * 100}, index=[0])], ignore_index=True)

    if df['10일'][i-1] > df['25일'][i-1] and df['10일'][i] < df['25일'][i] and stock_count > 0:  # 데드크로스 발생
        initial_money += df['종가'][i] * stock_count
        transactions = pd.concat([transactions, pd.DataFrame({'Date': df.index[i], 'Action': 'Sell', 'Price': df['종가'][i], 'Quantity': stock_count, 'Total': df['종가'][i] * stock_count}, index=[0])], ignore_index=True)

        stock_count = 0
        golden_cross = False

# 최종 수익 계산
final_asset = initial_money + (stock_count * df['종가'][-1])
profit = final_asset - 1000000000

print(transactions)

print(f"최종 자산: {final_asset:,.0f}원")
print(f"수익/손실: {profit:,.0f}원")

if profit > 0:
    print("이익입니다.")
else:
    print("손실입니다.")
```

```python
# prompt: 위의 전략을 유지하되, 다른 조건을 추가한다. 골든크로스 이후  매입시 2일전 주가가 1일전 주가대비 5% 내려가면 매입을 중단하고 데드크로스는 그 반대의 경우로 매도매수를 진행하라

# 이전 코드에서 필요한 부분만 가져와서 수정했습니다.

# 삼성전자 티커 심볼
ticker = '005930'  # company_info['삼성전자']를 직접 사용하는 것보다 명시적으로 ticker를 지정하는 것이 좋습니다.

# 2020년 1월 1일부터 현재까지의 주식 정보 가져오기
df = stock.get_market_ohlcv_by_date("20200101", "20240401", ticker)

# 10일, 25일 이동평균 계산
df['10일'] = df['종가'].rolling(window=10).mean()
df['25일'] = df['종가'].rolling(window=25).mean()

# 초기 자금
initial_money = 1000000000

# 주식 보유량
stock_count = 0

# 거래 내역 (데이터프레임 형태로 저장)
transactions = pd.DataFrame(columns=['Date', 'Action', 'Price', 'Quantity', 'Total'])

# 골든크로스, 데드크로스 판단 및 거래
golden_cross = False
for i in range(25, len(df)):
    if df['10일'][i-1] < df['25일'][i-1] and df['10일'][i] > df['25일'][i]: # 골든 크로스 발생
        golden_cross = True
        # 2일 전 주가가 1일 전 주가 대비 5% 하락했는지 확인
        if i >= 2 and df['종가'][i-2] < df['종가'][i-1] * 0.95:
            golden_cross = False # 매수 중단
            continue

    if golden_cross and initial_money >= df['종가'][i] * 100:
        stock_count += 100
        initial_money -= df['종가'][i] * 100
        transactions = pd.concat([transactions, pd.DataFrame({'Date': df.index[i], 'Action': 'Buy', 'Price': df['종가'][i], 'Quantity': 100, 'Total': df['종가'][i] * 100}, index=[0])], ignore_index=True)

    if df['10일'][i-1] > df['25일'][i-1] and df['10일'][i] < df['25일'][i] and stock_count > 0:  # 데드크로스 발생
        # 2일 전 주가가 1일 전 주가 대비 5% 상승했는지 확인
        if i >= 2 and df['종가'][i-2] > df['종가'][i-1] * 1.05:
            continue # 매도 중단

        initial_money += df['종가'][i] * stock_count
        transactions = pd.concat([transactions, pd.DataFrame({'Date': df.index[i], 'Action': 'Sell', 'Price': df['종가'][i], 'Quantity': stock_count, 'Total': df['종가'][i] * stock_count}, index=[0])], ignore_index=True)
        stock_count = 0
        golden_cross = False

# 최종 수익 계산
final_asset = initial_money + (stock_count * df['종가'][-1])
profit = final_asset - 1000000000

print(transactions)

print(f"최종 자산: {final_asset:,.0f}원")
print(f"수익/손실: {profit:,.0f}원")

if profit > 0:
    print("이익입니다.")
else:
    print("손실입니다.")
```
