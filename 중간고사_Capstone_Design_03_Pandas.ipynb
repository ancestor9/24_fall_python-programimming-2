```python
import pandas as pd

file_path = '/content/salaries_by_college_major.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
df.shape
```

```python
df['Undergraduate Major'].value_counts()
```

```python
# prompt: Undergraduate Major 별로 그룹핑해서

# Undergraduate Major 별로 그룹핑하여 평균 시작 급여, 중간 경력 급여 등을 확인합니다.
grouped_by_major = df.groupby('Undergraduate Major').agg({
    'Starting Median Salary': 'mean',
    'Mid-Career Median Salary': 'mean',
    'Mid-Career 10th Percentile Salary': 'mean',
    'Mid-Career 90th Percentile Salary': 'mean'
})

grouped_by_major.sort_values(by='Starting Median Salary', ascending=False)
```

```python
# prompt: 초봉이 가장 높은 학위는 무엇일까?

# 초봉이 가장 높은 학위 찾기
highest_starting_salary_major = df.loc[df['Starting Median Salary'].idxmax()]

print("초봉이 가장 높은 학위:")
print(highest_starting_salary_major['Undergraduate Major'])
```

```python
df.loc[df['Starting Median Salary'].idxmax()]
```

```python
# prompt: 대학 졸업 후 소득이 가장 낮은 전공은 무엇일까?

# 대학 졸업 후 소득이 가장 낮은 전공 찾기
lowest_starting_salary_major = df.loc[df['Starting Median Salary'].idxmin()]

print("대학 졸업 후 소득이 가장 낮은 전공:")
print(lowest_starting_salary_major['Undergraduate Major'])
```

```python
# prompt: 어떤 학위가 가장 높은 소득 잠재력을 가지고 있을까?

# 중간 경력 봉급이 가장 높은 학위 찾기
highest_mid_career_salary_major = df.loc[df['Mid-Career Median Salary'].idxmax()]

print("가장 높은 소득 잠재력을 가진 학위:")
print(highest_mid_career_salary_major['Undergraduate Major'])
```

```python
# prompt: 소득 관점에서 위험성이 가장 낮은 대학 전공은 무엇일까?

# 소득 관점에서 위험성이 가장 낮은 대학 전공 찾기
# 즉, 중간 경력 10%와 90% 백분위 수 급여의 차이가 가장 작은 전공을 찾습니다.

df['Spread'] = df['Mid-Career 90th Percentile Salary'] - df['Mid-Career 10th Percentile Salary']

lowest_spread_major = df.loc[df['Spread'].idxmin()]

print("소득 관점에서 위험성이 가장 낮은 대학 전공:")
print(lowest_spread_major['Undergraduate Major'])
```

```python
business_majors = df[df['Group'] == 'Business']['Undergraduate Major'].values.tolist()
hass_majors = df[df['Group'] == 'HASS']['Undergraduate Major'].values.tolist()
stem_majors = df[df['Group'] == 'STEM']['Undergraduate Major'].values.tolist()
```

```python
# STEM 학위의 평균 중간 경력 급여
stem_median_salary = df[df['Undergraduate Major'].isin(stem_majors)]['Mid-Career Median Salary'].mean()

# HASS 학위의 평균 중간 경력 급여
hass_median_salary = df[df['Undergraduate Major'].isin(hass_majors)]['Mid-Career Median Salary'].mean()

# 경영 학위의 평균 중간 경력 급여
business_median_salary = df[df['Undergraduate Major'].isin(business_majors)]['Mid-Career Median Salary'].mean()

print(f"STEM 학위의 평균 중간 경력 급여: {stem_median_salary}")
print(f"HASS 학위의 평균 중간 경력 급여: {hass_median_salary}")
print(f"경영 학위의 평균 중간 경력 급여: {business_median_salary}")

# 평균 중간 경력 급여가 가장 높은 학위 분야 출력
if stem_median_salary > hass_median_salary and stem_median_salary > business_median_salary:
  print("STEM 학위가 평균적으로 가장 높은 소득을 벌었습니다.")
elif hass_median_salary > stem_median_salary and hass_median_salary > business_median_salary:
  print("HASS 학위가 평균적으로 가장 높은 소득을 벌었습니다.")
elif business_median_salary > stem_median_salary and business_median_salary > hass_median_salary:
  print("경영 학위가 평균적으로 가장 높은 소득을 벌었습니다.")
else:
  print("평균 중간 경력 급여가 비슷합니다.")
```
