## Client에서 get 방식으로 데이터를 조회

```python
import requests

# ngrok에서 받은 public_url을 사용
public_url = "https://6218-34-86-83-103.ngrok-free.app/"
# https://6218-34-86-83-103.ngrok-free.app/

# 클라이언트가 10명의 사용자 데이터를 요청하는 예시
response = requests.get(f"{public_url}/users/10")

# 서버에서 받은 데이터를 출력
if response.status_code == 200:
    users = response.json()
    for user in users:
        print(user)
else:
    print(f"Failed to retrieve users: {response.status_code}")
```
